### Classificação

Para a classificação, o procedimento acima é realizado para todas as imagens da base de dados. É criada uma lista de tuplas chamada *data* (linhas 32 e 33 da classe Model). Cada tupla contém (i) imagem do rosto de uma pessoa e (ii) a classe a que aquela imagem pertence. Cada pessoa diferente pertence a uma classe.

As imagens são então fornecidas a um modelo, que procura aprender as características de cada rosto. O modelo usa validação cruzada, com k=3. 

Na validação cruzada, a base de dados é dividida em k pastas. São realizadas k iterações, sendo que a cada iteração uma pasta é selecionada como teste e as *k-1* pastas restantes são usadas para treino.

A cada iteração, o modelo deve classificar quem é quem nas imagens, a partir daquilo que aprendeu com as imagens de treino.

Os experimentos conduzidos avaliam o desempenho de quatro algoritmos de classificação: SVM, Naive Bayes, LDA e KNN com três vizinhos. As métricas levadas em consideração foram a *acurácia*, *precisão*, *revocação* e *f1-score*.

* Acurácia:

    Intuitivamente, é a taxa geral de acerto. É calculada dividindo todos os acertos pelo número total de perguntas.
    
    \begin{equation*}
    Acuracia = \frac{Acertos}{Acertos + Erros}
    \end{equation*}

* Precisão:

    Intuitivamente, pode ser vista como a seguinte pergunta: de todas as vezes que o classificador disse que tais objetos eram da classe $\alpha$, quantas vezes ele acertou?
    
    Formalmente, para um problema multiclasse, a precisão é calculada pela soma dos valores da diagonal principal de uma matriz de confusão (sendo a diagonal principal os acertos, ou *true positives*), dividindo-se pelo total de predições para aquela classe, sejam elas certas (*true positives*) ou erradas (*false positives*).
    
    \begin{equation*}
    Precisao = \frac{TP}{TP + FP}
    \end{equation*}

* Revocação:

    Intuitivamente, pode ser vista como a seguinte pergunta: se uma classe $\alpha$ tem *n* objetos, quantos desses *n* objetos foram identificados pelo classificador?
    
    Formalmente, para um problema multiclasse, a revocação é calculada pela soma dos valores da diagonal principal de uma matriz de confusão (sendo a diagonal principal os acertos), dividindo-se pelo total de instâncias daquela classe na base de dados (*true positives* e *false negatives*).
    
    \begin{equation*}
    Revocacao = \frac{TP}{TP + FN}
    \end{equation*}

* F1-Score:

    Intuitivamente, representa um *tradeoff* entre as precisão e revocação. 
    
    Formalmente, é a média harmônica entre precisão e revocação.
    
    \begin{equation*}
    F1score = 2 * \frac{precisao\ *\ revocacao}{precisao + revocacao}
    \end{equation*}
    

A célula seguinte executa os experimentos. São 26 pessoas (ou 26 classes) na base de dados, sendo cada uma identificada por uma letra do alfabeto. As tabelas abaixo apresentam os resultados de acurácia, precisão, revocação e f1-score para cada uma das classes em todos os algoritmos usados.

Cada um dos algoritmos foi treinado com as mesmas imagens, usando os mesmos atributos. Para gerar os atributos, a imagem equalizada de cada face (que é uma matriz 512x512) foi passada para o algoritmo *principal component analysis* (PCA), que é um algoritmo de redução de dimensionalidade.

Em resumo, o que o PCA faz é diminuir os atributos para um determinado número de componentes (no caso deste trabalho, 50 componentes). Ele faz isso convertendo um conjunto de variáveis possivelmente correlacionadas (e portanto redundantes) num conjunto linearmente independente. No caso deste trabalho, se a imagem inteira da face de cada pessoa fosse enviada para treinamento, os modelos seriam treinados com 262.144 atributos, o que além de ser custoso pode incluir muitas partes do rosto das pessoas que pouco contribuem para distinguir quem é quem. Com o PCA configurado para 50 componentes, ele reduzirá o número de atributos para 50.
